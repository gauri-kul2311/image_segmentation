```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy import ndimage
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from matplotlib.colors import hsv_to_rgb

img1 = cv2.imread('./corn1.jpg')
plt.imshow(img1)
plt.show()
```

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
plt.show()
```

```python
hsv_img1 = cv2.cvtColor(img1, cv2.COLOR_RGB2HSV)
plt.imshow(hsv_img1)
```

```python
lower_green1 = np.array([30,0,20])
upper_green1 = np.array([100,255,255])
mask1 = cv2.inRange(hsv_img1, lower_green1, upper_green1)
result1 = cv2.bitwise_and(img1, img1, mask=mask1)
plt.subplot(1, 2, 1)
plt.imshow(mask1, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(result1)
plt.show()
```

```python
lower_brown1 = np.array([10,0,10])
upper_brown1 = np.array([30,255,255])
disease_mask1 = cv2.inRange(hsv_img1, lower_brown1, upper_brown1)
disease_result1 = cv2.bitwise_and(img1, img1, mask=disease_mask1)
plt.subplot(1, 2, 1)
plt.imshow(disease_mask1, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(disease_result1)
plt.show()
```

```python
final_mask = mask1 + disease_mask1
final_result = cv2.bitwise_and(img1, img1, mask=final_mask)
plt.figure(figsize=(15,15))
plt.subplot(1, 2, 1)
plt.imshow(final_mask, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(final_result)
plt.show()
```
